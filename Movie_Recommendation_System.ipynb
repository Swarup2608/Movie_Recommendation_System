# **Importing of libraries**

```python
import numpy as np
import pandas as pd
import matplotlib as mt
import difflib
from  sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# **Data Collection**

```python
df = pd.read_csv('movies.csv')
```

```python
# Printing the head
df.head()
```

```python
# Getting the shape of the date
df.shape
```

```python
# Selecting features
selected_features = ['genres','keywords','tagline','cast','director','original_language','popularity','crew']
```

```python
for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
# combining all the  selected features

combined = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']+' '+df['original_language']+' '+str(df['popularity'])+' '+df['crew']
```

```python
combined
```

```python
# Text to vectors
# Iniating the vectorizer
Vectorizer = TfidfVectorizer()
```

```python
f_vector = Vectorizer.fit_transform(combined)
```

```python
print(f_vector)
```

# **Identifying the similarity between the features using COSINE Similarity**

```python
similarity = cosine_similarity(f_vector)
```

```python
similarity
```

# **Getting a movie name from the user**

```python
f_mve = input('Enter your favorite movie name: ')
```

```python
# Creating list of titles
l_titles = df['title'].tolist()
l_titles
```

```python
# Find the match to the given name
get_match = difflib.get_close_matches(f_mve,l_titles)
get_match
```

```python
close_match = get_match[0]
close_match
```

```python
index_mve = df[df.title == close_match]['index'].values[0]
index_mve
```

```python
# Getting list of similar movies
sim_scre = list(enumerate(similarity[index_mve]))
sim_scre
```

```python
# As the whole list of movies are printed
# Let sort the only similar ones
sorted_mves = sorted(sim_scre,key=lambda x:x[1],reverse=True)
sorted_mves
```

```python
len(sorted_mves)
```

```python
# Print similar movies
print('The recommended movies are: ')
i = 1 
for mve in sorted_mves[:30]:
    ind = mve[0]
    tit = df[df.index==ind]['title'].values[0]
    print(i,'.',tit)
    i+=1
```

# **THE MAIN SYSTEM**

```python
f_mve = input('Enter your favorite movie name: ')
get_match = difflib.get_close_matches(f_mve,l_titles)
close_match = get_match[0]
index_mve = df[df.title == close_match]['index'].values[0]
sim_scre = list(enumerate(similarity[index_mve]))
sorted_mves = sorted(sim_scre,key=lambda x:x[1],reverse=True)
print('The recommended movies for you because you liked',f_mve,': ')
i = 1 
for mve in sorted_mves[:30]:
    ind = mve[0]
    tit = df[df.index==ind]['title'].values[0]
    print(i,'.',tit)
    i+=1
```
